#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
salario_hora = (input("Qual seu salario por hora?"))
horas_contratadas = (input('Quantas horas por semana você trabalha?'))
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
salario_mensal = float(salario_hora) * (float(horas_contratadas)*4)
print(salario_mensal)
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
ir = 0.11 * salario_mensal
print (f'O seu desconto de IR será de R$ {ir}')
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
inss = 0.08 * salario_mensal
print (f'O seu desconto de INSS é de R${inss}')
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
sindicato = 0.05 * salario_mensal
print(f'O desconto para o sindicato é de R${sindicato}')
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
salario_liquido = salario_mensal - (float(ir)+float(inss)+float(sindicato))
print(f'Seu salario liquido é de R${salario_liquido}')
```

#### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
tamanho = float(input('Qual o tamanho do seu arquivo em MB?: '))
velocidade = float(input('Qual a velocidade da sua internet?: '))

megas = tamanho * 8
tempo = megas / velocidade
tempo_min = round(tempo / 60,2)

print(f'O tempo para baixar o arquivo será de {tempo_min} minutos')
```
